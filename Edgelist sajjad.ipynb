```python
import networkx as nx
import matplotlib.pylab as plt
g=nx.read_edgelist("discretenode.txt")
print(nx.info(g))
nx.draw(g)
plt.show()
# number of nodes
print("number of noeds:",nx.number_of_nodes(g))

#number of edges
print("number of edges:",nx.number_of_edges(g))

#average clustering
nx.average_clustering(g)

# diameter
print("Diameter:",nx.diameter(g))
# this diameter shows ...

# average degree
sum = 0
for n in g.nodes():
    sum = sum + g.degree(n)
print("Average degree:", sum/g.number_of_nodes())
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
from networkx.algorithms import community
from networkx.algorithms import bipartite
g = nx.read_edgelist("discretenode.txt")
c = bipartite.color(g)
nx.set_node_attributes(g, c, 'bipartite')
type1  = {n for n, d in g.nodes(data=True) if d['bipartite']==0}
type2  = {n for n, d in g.nodes(data=True) if d['bipartite']==1}
type1g = bipartite.projected_graph(g, type1)
type2g = bipartite.projected_graph(g, type2)
nx.draw(type1g, with_labels=True)
nx.draw(type2g, with_labels = True)
```
